8. Create a 2d array with 1 on border and 0 inside.

```python

def border_ones(n):
    a = np.zeros((n,n), dtype=int)
    a[0,:] = 1
    a[-1,:] = 1
    a[:,0] = 1
    a[:,-1] = 1
    return a

border_5 = border_ones(5)
print("5x5 border ones:\n", border_5)
```

#Problem- 4: Matrix Operations:

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
```

 1. Prove A.A−1 = I.

```python

A = np.array([[3,4],[7,8]], dtype=float)
A_inv = np.linalg.inv(A)
product_A_Ainv = A.dot(A_inv)
print("A @ A_inv:\n", product_A_Ainv)
print("Is close to identity? ->", np.allclose(product_A_Ainv, np.eye(2)))
```

2. Prove AB̸ = BA.

```python

B = np.array([[5,3],[2,1]], dtype=float)
AB = A.dot(B)
BA = B.dot(A)
print("A @ B:\n", AB)
print("B @ A:\n", BA)
print("Are AB and BA equal (elementwise)? ->", np.array_equal(AB, BA))
```

3. Prove (AB)T = BTAT.

```python

lhs = (A.dot(B)).T
rhs = B.T.dot(A.T)
print("(A@B).T:\n", lhs)
print("B.T @ A.T:\n", rhs)
print("Are they equal (allclose)? ->", np.allclose(lhs, rhs))
```

 • Solve the following system of Linear equation using Inverse Methods.
 2x −3y+z =−1
 x −y+2z =−3
 3x +y−z =9
 {Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
 • Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
 of Numpy}

```python

# Solve linear system
M = np.array([[2, -3, 1],
              [1, -1, 2],
              [3,  1, -1]], dtype=float)
C = np.array([-1, -3, 9], dtype=float)

# Solve with np.linalg.solve (preferred)
sol_solve = np.linalg.solve(M, C)

# Solve using inverse (explicit) to compare
M_inv = np.linalg.inv(M)
sol_inv = M_inv.dot(C)

print("Solution using np.linalg.solve:", sol_solve)
print("Solution using inverse:", sol_inv)
print("Are solutions equal? ->", np.allclose(sol_solve, sol_inv))
```
