```python
import pandas as pd
import numpy as np

df = pd.read_csv("Titanic-Dataset.csv")

df.info()
df.head()
```

# 3.1 Sorting and Subsetting:

Problem 1- Sorting:

```python
# 1. Create dataframe called fare that contains only Fare column
fare = df[['Fare']]
print("fare.head():")
print(fare.head())

# 2. DataFrame class_age containing Pclass and Age
class_age = df[['Pclass', 'Age']]
print("\nclass_age.head():")
print(class_age.head())

# 3. DataFrame survived_gender containing Survived and Sex
survived_gender = df[['Survived', 'Sex']]
print("\nsurvived_gender.head():")
print(survived_gender.head())
```

Problem- 2- Subsetting:

 Subsetting Rows:

```python
# 1. fare_gt_100 passengers with fare > 100
fare_gt_100 = df[df['Fare'] > 100]
print("fare_gt_100.shape:", fare_gt_100.shape)
print(fare_gt_100.head())

# 2. first_class passengers with Pclass = 1
first_class = df[df['Pclass'] == 1]
print("\nfirst_class.shape:", first_class.shape)
print(first_class.head())

# 3. female_under_18 Age < 18 and Sex = female
female_under_18 = df[(df['Age'] < 18) & (df['Sex'] == 'female')]
print("\nfemale_under_18.shape:", female_under_18.shape)
print(female_under_18.head())
```

Subsetting Rows by Categorical variables:

```python
# 1. embarked_c_or_s
embarked_c_or_s = df[df['Embarked'].isin(['C','S'])]
print("embarked_c_or_s.shape:", embarked_c_or_s.shape)
print(embarked_c_or_s.head())

# 2. first_second_class Pclass in [1,2]
first_second_class = df[df['Pclass'].isin([1,2])]
print("\nfirst_second_class.shape:", first_second_class.shape)
print(first_second_class.head())
```

# 3.2 ExploratoryDataAnalysis

Which passenger had the highest fare paid relative to their age?

```python
# Fill Age with median
df_age_filled = df.copy()
median_age = df_age_filled['Age'].median()
df_age_filled['Age'] = df_age_filled['Age'].fillna(median_age)

# 1. Add column fare_per_year  Fare / Age
df_age_filled['fare_per_year'] = df_age_filled['Fare'] / df_age_filled['Age']

# 2. Subset where fare_per_year hugher than 5
high_fare_age = df_age_filled[df_age_filled['fare_per_year'] > 5]

# 3. Sort descending by fare_per_year
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

# 4. Select Name and fare_per_year
result = high_fare_age_srt[['Name', 'fare_per_year']]

# 5. Show result
print(result.head(10))
```

Which adult male passenger(age≥18 and Sexis’male’)paid the highest fare relative to their
 class?

```python
# 1. Add column fare_per_class  Fare / Pclass
df_age_filled['fare_per_class'] = df_age_filled['Fare'] / df_age_filled['Pclass']

# 2. Subset adult males: Age >= 18 and Sex == 'male'
adult_males = df_age_filled[(df_age_filled['Age'] >= 18) & (df_age_filled['Sex'] == 'male')]

# 3. Sort descending by fare_per_class
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

# 4. Select Name, Age, fare_per_class
result_adult_males = adult_males_srt[['Name','Age','fare_per_class']]

# 5. View result (top rows)
print(result_adult_males.head(10))
```

# 3.3 Exploratory Data Analysis with Group-by Method

 What percent of the total fare revenue came from each passenger class?

```python
# 1. total fare of all passengers
total_fare = df['Fare'].sum()
print("Total fare:", total_fare)

# 2 and 3. total fare per class using groupby (safer and concise)
fare_by_class = df.groupby('Pclass')['Fare'].sum().sort_index()  # index 1,2,3
print("\nFare by class:\n", fare_by_class)

# 4. Combine totals into a list
totals_list = fare_by_class.tolist()

# 5. Proportion
fare_pct_by_class = (fare_by_class / total_fare) * 100
print("\nPercent of total fare revenue by class:\n", fare_pct_by_class.round(2))
```

 What percent of the total number of passengers on the Titanic belonged to each age group
 (e.g., child, adult, senior)?

```python
df2 = df_age_filled.copy()

# 1. Create age_group column
def age_group_label(age):
    if age < 18:
        return 'child'
    elif age >= 65:
        return 'senior'
    else:
        return 'adult'

df2['age_group'] = df2['Age'].apply(age_group_label)

# 2. Total passengers
total_passengers = len(df2)

# 3. Count per age group
counts = df2['age_group'].value_counts().sort_index()
print("Counts:\n", counts)

# 4. Proportion as percentage
pct_by_age_group = (counts / total_passengers) * 100
print("\nPercent by age group:\n", pct_by_age_group.round(2))
```
